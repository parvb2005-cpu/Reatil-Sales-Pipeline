## Answer 10

```python
# Now we are going to find the top 3 best-selling products by total quantity by writing a sql query.
query = """
SELECT
    product_id,
    product_name,
    SUM(quantity) AS total_quantity_sold
FROM retail_sales
GROUP BY product_id, product_name
ORDER BY total_quantity_sold DESC
LIMIT 3;
"""
# Now we are going to execute the query and store the result in our dataframe.
top_3_products = pd.read_sql_query(query, connection)

# Now we are going to print the result
top_3_products
```

## Answer 11

```python
# Now ae are going to write a SQL query to find total revenue per store per day from the 'retail_sales' table.

query = """
SELECT
    store_id,
    store_name,
    sale_date,
    SUM(total_revenue) AS total_revenue
FROM retail_sales
GROUP BY store_id, store_name, sale_date
ORDER BY store_id, sale_date;
"""

# Now we are going to execute the query as well as we will print it.
revenue_per_store = pd.read_sql_query(query, connection)
print(revenue_per_store)
```

```python
# Now we are going define a function called 'run_pipeline' that will runs all the above steps together.
def run_pipeline():
    # Load files
    df1 = pd.read_csv("C:\\Users\\Admin\\Downloads\\sales_data.csv")
    df2 = pd.read_csv("C:\\Users\\Admin\\Downloads\\products.csv")
    df3 = pd.read_csv("C:\\Users\\Admin\\Downloads\\stores.csv")

    print("Files loaded successfully.")

    # Data Cleaning
    duplicate_count = df1.duplicated().sum()
    df1 = df1.drop_duplicates()

    df1["quantity"] = df1["quantity"].fillna(0)
    df1 = df1.dropna(subset=["amount"]).copy()

    print(f"Duplicates Removed : {duplicate_count}")
    print("Data cleaned successfully.")

    # Merging DataFrames together
    merged_df = pd.merge(df1, df2, on="product_id", how="inner")
    final_df = pd.merge(merged_df, df3, on="store_id", how="inner")

    # Transforming Data
    final_df["total_revenue"] = final_df["quantity"] * final_df["price"]

    print("Transformation completed.")

    # Final Step Loading DataFrame into a Database
    connection = sqlite3.connect("reatil_sales.db")
    final_df.to_sql("retail_sales", connection, if_exists="replace", index=False)

    print("Data loaded successfully into SQLite database.")
    print("Pipeline executed successfully!")

    return final_df
```

```python
final_df = run_pipeline()
```
